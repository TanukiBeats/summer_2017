**Describe the following**

Supervised vs unsupervised (chpt 2)
supervised - a statistical model which gives an output from given inputs.
unsupervised - a statistical model which has inputs but no supervising output. It is unsupervised becuase there is no response to the model with which we can evalute the accuracy of the model. One example is cluster analysis. 

Parametric vs non-parametric (chpt 2) - in general, statistical learning methods for finding f(x) can be classified into these two categories.
parametric - estimating f(x) becuase an issue for estimating a set of parameters once a assumption is made for the functional form of f(x). It tends to be less accurate and can lead of overfitting of data if too flexible of a model is used to overcompensate for the lack of accuracy. 
non-parametric - no assumptions are made about the functional form of f(x). An accurate estimate is made, without overfitting the data or causing roughness. However, they it is a much more laborious procedure compared to parametric methods. 

Prediction vs inference (chpt 2) - two of the main reasons that we would want to find f(x).
prediction - the inputs are known but the output is not necessarily known. f(x) is used to predict values for the output. The exact form of f(x) does not need to be known.
inference - finding the relationship between the input variables of f(x) and the output. Often looking for individual relationships between the inputs and output, as well as which inputs have the most significant impact. 

Bias vs variance (and their tradeoff) (chpt 2) - both are *inherently* non-negitive quantities. The bias-variance tradeoff is an equation which relates the variance, bias, and variance in error of f(x). The challenge is in finding a tradeoff where varience and bias are *both* low. 
bias - more attributable to human error. It mainly comes from assumptions made about some phenomenom.  
variance - how much f(x) would be different if a different training data set was used for the model. More flexible statistical methods tend to have higher variance. 

What are our sources of reducible and irreducible error when constructing a model? - 
reducible error comes from the way in which f(x) is found. It can be improved by using a different statistical method. 
irreducible error - comes from the undetermined constant. The constant is not related to the input variables, and thus cannot be improved upon with statistical methods. Comes from the dataset.

Quantative vs. qualitative (categorical) - the two types of variables.
quantitative - variables which are numerical values. Regression problems have a quantative responce. 
qualitative variables - have the value of some category or class. Classification problems have a qualitative response. 
However, quantative and qualitative variables are not tied to the respect type of statistic problem. 

mean squared error - a measure for how close an observation from a model is close to the true response value. MSE is smaller as the predicited responses are more accurate. Training MSE is typically smaller than the test MSE.

expected test MSE - the average test MSE which would be obtained from repeatedly estimating f(x) using a large number of training sets each tested at xsub0

**understand the differences in these three measurements of error:**

$RSE=\sqrt{\frac{RSS}{n-2}}=\sqrt{\frac{\sum_{i=1}^n(y_i-\hat{y}_i)^2}{n-2}}$
RSE - *residual standard error*. A quantity, used along with the R squared statistic which asseses the quality of a linear regression fit. It is a measure of the standard deviation of the E (error term). The average amount that the response of the model will deviate from the true model. Measure of the lack of fit.

$RSS=\sum_{i=1}^n(y_i-\hat{y}_i)^2$
RSS - *residual sum of squares*. The difference between the ith observed response value and the ith response value that is predicted by the linear model

$R^2=\frac{TSS-RSS}{TSS}=\frac{\sum_{i=1}^n(y_i-\overline{y}_i)^2-\sum_{i=1}^n(y_i-\hat{y}_i)^2}{\sum_{i=1}^n(y_i-\overline{y}_i)^2}$
An alternative measure of fit. RSE is measured in units of "Y", so it can be unclear what the problem is. R squared is a proportion. R squared measures the proportion of variability in Y that can be explained using X. The closer the value is to 1, the more of the error can be explained by the regression. The closer the value is to 0, it means that the regression itself is not wrong but there could be a problem with the model or the inherenent error.

````python
import matplotlib.pyplot as plt
from sklearn import preprocessing
import numpy as np
from sklearn import datasets, linear_model
import pandas as pd

%matplotlib inline

#model: Y = 3X + 4 
pts=25
x=np.linspace(-50,50,num=pts)
valid_x=np.linspace(-50,50,num=15)
B0=4
B1=3
yactual=B0+B1*x
valid_y=B0+B1*valid_x
np.random.seed(123)

#add noise scaled to 25% of range to data
yrand=yactual+.25*(yactual.max()-yactual.min())*np.random.normal(size=pts)
valid_yrand=valid_y+.25*(valid_y.max()-valid_y.min())*np.random.normal(size=15)

for i in range(0,100,1):
    #calculate the testing data with i% strength of error
    valid_yrand=valid_y+float(i)/100*(valid_y.max()-valid_y.min())*np.random.normal(size=15)
    
    #plot the MSE between the actual data and the testing data
    plt.plot(i,((valid_y-valid_yrand)**2).mean(axis=None),marker='*')

plt.ylabel('MSE')
plt.xlabel('Percent Error in Testing Data')
````

````python
from math import sqrt
from scipy import stats
#model: Y = 3X + 4 
pts=25
x=np.linspace(-50,50,num=pts)
valid_x=np.linspace(-50,50,num=15)
B0=4
B1=3
yactual=B0+B1*x
valid_y=B0+B1*valid_x
np.random.seed(123)

#add noise scaled to 25% of range to data
yrand=yactual+.25*(yactual.max()-yactual.min())*np.random.normal(size=pts)

plt.plot(x,yactual,color='red',lw='3')
plt.scatter(x,yrand)
regr=linear_model.LinearRegression()

regr.fit(x.reshape(-1,1),yrand)
print('B0, B1: ',regr.intercept_, regr.coef_[0])
plt.plot(x,yactual,color='red',lw='3',label='actual')
plt.plot(x,regr.predict(x.reshape(-1,1)),ls='--',label='fit')
plt.legend(loc='lower right')

#calculate the RSE, SSR, $R^2$, 

#RSS
RSS = ((regr.predict(valid_x.reshape(-1,1))-valid_y)**2).mean(axis=None)
print('RSS =', RSS)

#RSE
RSE = (sqrt(RSS/(len(valid_x)-2)))
print('RSE =', RSE)

#R2
TSS = ((valid_y-regr.predict(valid_x.reshape(-1,1)).mean(axis=None))**2).mean(axis=None)
R2 = (TSS-RSS)/TSS
print('R2 =', R2)
print('scikit regression score=', regr.score(valid_x.reshape(-1,1), valid_y))
  
#95% confidence intervals for both 
#$\beta_i$, and evaluate the P-value at 0.05 significance level for $\beta_1$

#SE for B0 and B1
SEB0 = RSE**2*((1/len(valid_x))+valid_y.mean(axis=None)**2/np.sum((valid_y-valid_y.mean\
            (axis=None))**2))
CONF0 = [B0-2*SEB0,B0+2*SEB0]
print('Confidence interval for B0=', CONF0)

SEB1 = (RSE**2/np.sum((valid_y-valid_y.mean(axis=None))**2))
CONF1 = [B1-2*SEB1,B1+2*SEB1]
print('Confidence interval for B1=', CONF1)

#t for B1
t = (B1-0)/float(SEB1)

#the p-value tells us the propbability of achieving a value
#at least as extreme as the one for our dataset if the null
#hypothesis were true
p=stats.t.sf(np.abs(t),len(valid_x)-1)*2 #multiply by two for two-sided test
print('p', p)
````
